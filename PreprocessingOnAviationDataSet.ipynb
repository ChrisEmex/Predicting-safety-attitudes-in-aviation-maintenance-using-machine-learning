```python
import pandas as pd

# Assuming your data is already loaded into the 'data' DataFrame

# Create an empty DataFrame to hold the results
final_result = pd.DataFrame()

# Iterate through each column in the data
for column in data.columns:
    # Calculate the value counts for each column (1, 2, 3, 4, 5 answers)
    value_counts = data[column].value_counts().sort_index()  # Sort index to have 1, 2, 3, 4, 5 in order
    # Ensure all answers (1, 2, 3, 4, 5) are present, even if their count is zero
    value_counts = value_counts.reindex([1, 2, 3, 4, 5], fill_value=0)
    
    # Add the value counts as a new column in the final result
    final_result[column] = value_counts

# Save the result to a single Excel file
final_result.to_excel('column_value_counts_reshaped.xlsx', index_label="Answer")

print("All value counts saved to 'column_value_counts_reshaped.xlsx'")
```

```python
# Create an empty dictionary to hold results
final_result = {}

# Iterate through each column in the data
for column in data.columns:
    # Get unique values dynamically
    unique_values = sorted(data[column].dropna().unique())  # Ensure values are sorted
    # Calculate value counts
    value_counts = data[column].value_counts().reindex(unique_values, fill_value=0)
    # Store in dictionary
    final_result[column] = value_counts

# Convert dictionary to DataFrame
final_result_df = pd.DataFrame(final_result)

# Save to Excel
final_result_df.to_excel("column_value_counts_reshaped.xlsx", index_label="Answer")

print("All value counts saved to 'column_value_counts_reshaped.xlsx'")
```
